```python
# # AQI Calculation
# #Goal: Compute AQI from pollutant concentrations (pm25, o3).Use the US EPA AQI formula (simplified here for PM2.5 and O3):
# def calculate_aqi(pm25, o3):
#     # Sub-index for PM2.5 (example; adjust breakpoints to your country's standard)
#     if pm25 <= 12:
#         pm25_ratio = (pm25 / 12) * 50
#     elif pm25 <= 35.4:
#         pm25_ratio = 51 + ((pm25 - 12.1) / (35.4 - 12.1)) * 49
#     else:
#         pm25_ratio = 101 + ((pm25 - 35.5) / (250 - 35.5)) * 149  # Simplified

#     # Sub-index for O3 (8-hour average; adjust thresholds)
#     if o3 <= 54:
#         o3_ratio = (o3 / 54) * 50
#     elif o3 <= 70:
#         o3_ratio = 51 + ((o3 - 55) / (70 - 55)) * 49
#     else:
#         o3_ratio = 101 + ((o3 - 71) / (100 - 71)) * 149  # Simplified

#     # AQI = max of sub-indices
#     aqi = max(pm25_ratio, o3_ratio)
#     return round(aqi)

# merged_df['AQI'] = merged_df.apply(lambda x: calculate_aqi(x['pm25'], x['o3']), axis=1)

def calculate_aqi_nepal(pm25, o3):
    def get_pm25_sub_index(pm):
        # Based on Nepal/India AQI categories for PM2.5 (24-hour average)
        breakpoints = [
            (0, 30, 0, 50),
            (31, 60, 51, 100),
            (61, 90, 101, 200),
            (91, 120, 201, 300),
            (121, 250, 301, 400),
            (251, 500, 401, 500)
        ]
        for (clow, chigh, ilow, ihigh) in breakpoints:
            if clow <= pm <= chigh:
                return ((ihigh - ilow) / (chigh - clow)) * (pm - clow) + ilow
        return 500  # Beyond measurable range

    def get_o3_sub_index(o):
        # Based on Indian standard for O3 (8-hour average, in µg/m³)
        breakpoints = [
            (0, 50, 0, 50),
            (51, 100, 51, 100),
            (101, 168, 101, 200),
            (169, 208, 201, 300),
            (209, 748, 301, 400),
            (749, 1000, 401, 500)
        ]
        for (clow, chigh, ilow, ihigh) in breakpoints:
            if clow <= o <= chigh:
                return ((ihigh - ilow) / (chigh - clow)) * (o - clow) + ilow
        return 500  # Beyond measurable range

    pm25_index = get_pm25_sub_index(pm25)
    o3_index = get_o3_sub_index(o3)

    return round(max(pm25_index, o3_index))
merged_df['AQI'] = merged_df.apply(lambda x: calculate_aqi_nepal(x['pm25'], x['o3']), axis=1)


```

```python
# feature engenearing(creating meaningful feature for the model)
# Extract time-based features before dropping 'date_and_time'
merged_df['hour'] = merged_df['date_and_time'].dt.hour
merged_df['day_of_week'] = merged_df['date_and_time'].dt.dayofweek
merged_df['month'] = merged_df['date_and_time'].dt.month

# Lag features (e.g., previous 3-hour AQI)
merged_df['AQI_lag_3'] = merged_df['AQI'].shift(3)

# Weather interaction terms
merged_df['temp_humidity'] = merged_df['temperature'] * merged_df['relative_humidity']
merged_df['wind_rain'] = merged_df['wind_speed'] * merged_df['rain']

# Now you can safely drop the 'date_and_time' column
#merged_df.drop('date_and_time', axis=1, inplace=True)
```

```python
# again data clening and prepared data fro modeling
# Handle missing values
merged_df.fillna(method='ffill', inplace=True)  # Forward-fill for time-series

# Drop duplicates
merged_df.drop_duplicates(inplace=True)

# Remove outliers in AQI ( beyond 3σ)
mean = merged_df['AQI'].mean()
std = merged_df['AQI'].std()
merged_df = merged_df[(merged_df['AQI'] <= mean + 3*std)]
```

```python
# train and test
from sklearn.model_selection import train_test_split

X = merged_df.drop(['AQI', 'pm25', 'o3','date_and_time'], axis=1)  # Drop raw pollutants (AQI is the target)
y = merged_df['AQI']

# Split with time-series awareness (e.g., last 20% as test)
test_size = int(0.2 * len(X))
X_train, X_test = X[:-test_size], X[-test_size:]
y_train, y_test = y[:-test_size], y[-test_size:]
```

```python
# model building
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Initialize model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print(f"MAE: {mean_absolute_error(y_test, y_pred)}")
print(f"R²: {r2_score(y_test, y_pred)}")


```

```python
# advance feature ingenearing
# time-serice feature
import pandas as pd
import numpy as np

# Ensure data is sorted by time
merged_df = merged_df.sort_values('date_and_time').reset_index(drop=True)

# Time-based features
merged_df['hour'] = merged_df['date_and_time'].dt.hour
merged_df['day_of_week'] = merged_df['date_and_time'].dt.dayofweek
merged_df['month'] = merged_df['date_and_time'].dt.month

# Lagged AQI (past 3, 6, 24 hours)
for lag in [3, 6, 24]:
    merged_df[f'AQI_lag_{lag}'] = merged_df['AQI'].shift(lag)

# Rolling statistics (24-hour window)
merged_df['rolling_temp_24h'] = merged_df['temperature'].rolling(24).mean()
merged_df['rolling_wind_24h'] = merged_df['wind_speed'].rolling(24).mean()
merged_df['rolling_humidity_24h'] = merged_df['relative_humidity'].rolling(24).mean()

# Exponential moving averages (EMA)
merged_df['ema_AQI_12h'] = merged_df['AQI'].ewm(span=12).mean()

# Drop rows with NaN from lag/rolling features
merged_df = merged_df.dropna()
```

```python
# Train-Test Split with TimeSeriesSplit
from sklearn.model_selection import TimeSeriesSplit

# Split data into features (X) and target (y)
X = merged_df.drop(['AQI', 'date_and_time', 'pm25', 'o3'], axis=1)
y = merged_df['AQI']

# TimeSeriesSplit (preserve temporal order)
tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
# lstm impementation
# data prepearation for lstm
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Scale data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Reshape into [samples, timesteps, features]
look_back = 24  # Use 24 hours of historical data
X_lstm = []
y_lstm = []
for i in range(len(X_scaled) - look_back):
    X_lstm.append(X_scaled[i:i+look_back, :])
    y_lstm.append(y[i+look_back])
X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)

# Train-test split for LSTM
train_size = int(0.8 * len(X_lstm))
X_train_lstm, X_test_lstm = X_lstm[:train_size], X_lstm[train_size:]
y_train_lstm, y_test_lstm = y_lstm[:train_size], y_lstm[train_size:]
```

```python
# build and train lstm model
model_lstm = Sequential()
model_lstm.add(LSTM(64, return_sequences=True, input_shape=(look_back, X_train_lstm.shape[2])))
model_lstm.add(Dropout(0.2))
model_lstm.add(LSTM(32))
model_lstm.add(Dense(1))

model_lstm.compile(optimizer='adam', loss='mse')
history = model_lstm.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=64,
                        validation_split=0.1, verbose=1)
```

```python
# Model Comparison & Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error

def evaluate_model(model, X_test, y_test):
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    return mae, rmse

# Evaluate LSTM
lstm_mae, lstm_rmse = evaluate_model(model_lstm, X_test_lstm, y_test_lstm)

# Evaluate FCNN
fcnn_mae, fcnn_rmse = evaluate_model(model_fcnn, X_test_scaled, y_test)

print(f"LSTM: MAE={lstm_mae:.2f}, RMSE={lstm_rmse:.2f}")
print(f"FCNN: MAE={fcnn_mae:.2f}, RMSE={fcnn_rmse:.2f}")

```
