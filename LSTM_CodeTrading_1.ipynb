```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Specify the ticker symbol
ticker_symbol = "AAPL"  # Replace with the desired ticker symbol

try:
    # Fetch the Ticker object
    ticker = yf.Ticker(ticker_symbol)

    # Attempt to retrieve market information
    ticker_info = ticker.info
    if "regularMarketPrice" in ticker_info:
        market_price = ticker_info['regularMarketPrice']
        print(f"Regular Market Price for {ticker_symbol}: {market_price}")
    else:
        print("Regular Market Price not found in the `info`. Attempting fallback...")
        market_price = None

    # Fallback: Retrieve historical data
    data = ticker.history(period="1mo", interval="1d")  # Fetch the last month's daily data
    if not data.empty:
        latest_close = data['Close'].iloc[-1]
        print(f"Fallback: Latest Close Price for {ticker_symbol}: {latest_close}")

        # Save data to a CSV file
        csv_file = f"{ticker_symbol}_historical_data.csv"
        data.to_csv(csv_file)
        print(f"Historical data saved to {csv_file}")

        # Visualize the data
        plt.figure(figsize=(10, 6))
        plt.plot(data['Close'], label="Close Price", color="blue", marker='o')
        plt.title(f"{ticker_symbol} - Historical Close Prices")
        plt.xlabel("Date")
        plt.ylabel("Close Price (USD)")
        plt.grid()
        plt.legend()
        plt.tight_layout()

        # Save the visualization
        image_file = f"{ticker_symbol}_historical_plot.png"
        plt.savefig(image_file)
        print(f"Visualization saved to {image_file}")
        plt.show()
    else:
        print("No historical data available for fallback.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import yfinance as yf
import pandas as pd

# List of stock tickers
tickers = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA']

# Define screening criteria
pe_ratio_threshold = 20
dividend_yield_threshold = 2  # Percent

# Initialize an empty list to store the filtered stocks
filtered_stocks = []

for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    try:
        pe_ratio = info.get('trailingPE', None)
        dividend_yield = info.get('dividendYield', None) * 100 if info.get('dividendYield') else None

        # Check if the stock meets the criteria
        if pe_ratio and dividend_yield:
            if pe_ratio < pe_ratio_threshold and dividend_yield >= dividend_yield_threshold:
                filtered_stocks.append({
                    'Ticker': ticker,
                    'P/E Ratio': pe_ratio,
                    'Dividend Yield': dividend_yield
                })
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# Display filtered stocks
filtered_df = pd.DataFrame(filtered_stocks)
print(filtered_df)
```
